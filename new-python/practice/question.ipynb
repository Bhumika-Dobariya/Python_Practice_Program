```python
class Bank_Account():
    def CreateAccount(self):
        self.name = input("enter name")
        self.accno = int(input("enter account number"))
        self.balance = int(input("enter initial balance"))
        
    def Display_Account(self):
        print("name",self.name)
        print("account number",self.accno)
        print("balance",self.balance)
        
    def Deposit(self):
        amount = int(input("enter deposit"))
        amount+=self.balance 
        print("your total amount is",amount)  
        
    def Withdraw(self):
        withdraw = int(input("enter amount")) 
        if self.balance>=withdraw:
              self.balance-=withdraw
              print("withdraw",withdraw)     
        else:
            print("your amount is insufficient")
    
    def show(self):
        print("available balance",self.balance)  
              
b1 = Bank_Account()
b1.CreateAccount()
b1.Display_Account()
b1.Deposit()
b1.Withdraw()
b1.show()
```

```python
class Person():
    def information(self):
        self.name = input("enter name")
        self.country = input("enter country")
        self.birthdate = int(input("date of birth"))
        self.birthmonth = int(input("month of birth"))
        self.birthyear = int(input("year of birth"))
        
    def display(self):
        print("enter name",self.name)
        print("enter country",self.country)
        print("enter birthdate",self.birthdate)
        print("enter birthmonth",self.birthmonth)
        print("enter birthyear",self.birthyear)
    
    def Cal_Age(self):
        self.current_date = int(input("enter current date"))
        self.current_month = int(input("enter current month"))
        self.current_year = int(input("enter current year"))
        
        self.year =self.current_year-self.birthyear
        self.month = self.current_month-self.birthmonth
        self.date = self.current_date- self.birthdate
    def show(self):
        print(f"you are{self.year} year {self.month} month {self.date} days old")
        
        
p1 = Person()
p1.information()
p1.display()
p1.Cal_Age() 
p1.show()       
        
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = {}
        
    def add_item(self,item,price):
        if item in self.items:
            self.items[item]+=price
        else:
            self.items[item]=price
   
    def remove_item(self,item):
        if item in self.items:
            del self.items[item]
        else:
            print(f"{item} is not in the shopping cart.")
         
    def calculate_total(self):
        total_price = sum(self.items.values())
        return total_price

s1 = ShoppingCart()
s1.add_item("apple",50)
s1.add_item("mango",25)
s1.add_item("apple",50)
s1.add_item("banana",50)

print("total price", s1.calculate_total())

s1.remove_item("banana")
s1.remove_item("orange")

print("total price", s1.calculate_total())           
            
```
